# Imports

```python
import json
import matplotlib.pyplot as plt

with open("europe.json", "r") as file:
    europe_json = json.load(file)
with open("world.json", "r") as file:
    world_json = json.load(file)


# Function to extract year and percentage from JSON
def extract_data(json_data):
    years = []
    percentages = []
    for entry in json_data:
        years.append(entry["dataYear"])
        percentages.append(entry["answersPercentage"][0]["value"])
    return years, percentages


# Extract data for Europe and World
europe_years, europe_percent = extract_data(europe_json)
world_years, world_percent = extract_data(world_json)
```

# Plots

```python
import matplotlib.pyplot as plt

# Set the font and font size
plt.rcParams["font.family"] = "Avenir Next"
plt.rcParams["font.size"] = 14

# Create the plot
plt.figure(figsize=(960 / 72, 1080 / 72))
plt.plot(europe_years, europe_percent, marker="o", label="Europe", color="magenta")
plt.plot(world_years, world_percent, marker="s", label="World", color="blue")

# Add title and labels
plt.title(
    "ITU - Percentage of Individuals Using the Internet (2005-2024)", fontsize=18
)  # Increase title size to 18
plt.xlabel("Year", fontsize=16)
plt.ylabel("Percentage of Individuals (%)", fontsize=16)
plt.legend()

# Improve x-axis ticks
plt.xticks(range(2005, 2025, 2))

# Add grid for better readability
plt.grid(True, linestyle="--", alpha=0.7)

# Add data labels for the last point of each series
plt.annotate(
    f"{europe_percent[-1]}%",
    xy=(europe_years[-1], europe_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="magenta",
)

plt.annotate(
    f"{world_percent[-1]}%",
    xy=(world_years[-1], world_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="blue",
)

# Adjust layout to prevent cutting off labels
plt.tight_layout()


# Show the plot
plt.show()
```
