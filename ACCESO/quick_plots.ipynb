# Imports

```python
import json
import matplotlib.pyplot as plt

with open("europe.json", "r") as file:
    europe_json = json.load(file)
with open("world.json", "r") as file:
    world_json = json.load(file)
with open("americas.json", "r") as file:
    americas_json = json.load(file)
with open("asia.json", "r") as file:
    asia_json = json.load(file)
with open("arab.json", "r") as file:
    arab_json = json.load(file)


# Function to extract year and percentage from JSON
def extract_data(json_data):
    years = []
    percentages = []
    for entry in json_data:
        years.append(entry["dataYear"])
        percentages.append(entry["answer"][0]["value"])
    return years, percentages


# Extract data for Europe and World
europe_years, europe_percent = extract_data(europe_json)
world_years, world_percent = extract_data(world_json)
americas_years, americas_percent = extract_data(americas_json)
asia_years, asia_percent = extract_data(asia_json)
arab_years, arab_percent = extract_data(arab_json)
```

# Plots

```python
import matplotlib.pyplot as plt

# Set the font and font size
plt.rcParams["font.family"] = "Avenir Next"
plt.rcParams["font.size"] = 12

# Create the plot
plt.figure(figsize=(960 / 72, 1080 / 72))
plt.plot(europe_years, europe_percent, marker="o", label="Europe", color="magenta")
plt.plot(americas_years, americas_percent, marker="^", label="Americas", color="orange")
plt.plot(asia_years, asia_percent, marker="x", label="Asia", color="green")
plt.plot(arab_years, arab_percent, marker="*", label="Arab States", color="red")
plt.plot(world_years, world_percent, marker="s", label="World", color="blue")

# Add title and labels
plt.title(
    "ITU - Percentage of Individuals Using the Internet (2005-2024)", fontsize=22
)  # Increase title size to 18
plt.xlabel("Year", fontsize=18)
plt.ylabel("Percentage of Individuals (%)", fontsize=18)
plt.legend()

# Improve x-axis ticks
plt.xticks(range(2004, 2026, 2))

# Add grid for better readability
plt.grid(True, linestyle="--", alpha=0.7)

# Add data labels for the last point of each series
plt.annotate(
    f"{europe_percent[-1]}%",
    xy=(europe_years[-1], europe_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="magenta",
)
plt.annotate(
    f"{americas_percent[-1]}%",
    xy=(americas_years[-1], americas_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="orange",
)

plt.annotate(
    f"{asia_percent[-1]}%",
    xy=(asia_years[-1], asia_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="green",
)

plt.annotate(
    f"{arab_percent[-1]}%",
    xy=(arab_years[-1], arab_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="red",
)

plt.annotate(
    f"{world_percent[-1]}%",
    xy=(world_years[-1], world_percent[-1]),
    xytext=(5, 5),
    textcoords="offset points",
    ha="left",
    va="bottom",
    color="blue",
)

# Adjust layout to prevent cutting off labels
plt.tight_layout()


# Show the plot
plt.show()
```
